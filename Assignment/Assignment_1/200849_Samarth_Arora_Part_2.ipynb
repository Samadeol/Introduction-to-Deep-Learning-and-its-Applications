Importing all the dependencies and the modules


```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

Reading CSV file and Plotting the averages city wise


```python
M = pd.read_csv('House_prediction.csv')
temp = list(M.head())
data = M.groupby(["city"])[temp[2:]]
data = data.mean()
print(len(data.head(5)))
for i in range(len(data.head(8))):
    plt.figure(figsize=(21, 9))
    x = data.index.to_list()
    y = data.loc[:,list(data.head(0))[i]]
    plt.plot(x,y)
    plt.ylabel(list(data.head(0))[i])
    plt.show()
data
```

Correlation plot using heatmaps, Higher density implies 


```python
sns.set(rc = {'figure.figsize':(10,10)})
sns.heatmap(data.corr(), annot=True)
```
